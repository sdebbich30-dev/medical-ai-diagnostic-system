```python
import matplotlib.pyplot as plt
import os
from PIL import Image

# Chemin des images
images_dir = '../data/raw'

# Lister les images
all_images = os.listdir(images_dir)
image_files = [f for f in all_images if f.endswith('.png') or f.endswith('.jpg')]

print(f"Images trouv√©es: {len(image_files)}")

# Afficher les 10 premi√®res images
plt.figure(figsize=(20, 8))
for i in range(min(10, len(image_files))):
    plt.subplot(2, 5, i+1)
    img_path = os.path.join(images_dir, image_files[i])
    img = Image.open(img_path)
    plt.imshow(img, cmap='gray')
    plt.title(image_files[i], fontsize=8)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
